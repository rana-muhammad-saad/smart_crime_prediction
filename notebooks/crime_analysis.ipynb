```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("C:/Users/BHATTI Computers/Desktop/smart_crime_prediction/data/processed/cleaned_crime_data.csv")

# Top 10 crime types
plt.figure(figsize=(8,6))
sns.countplot(data=df, y="crime_type", order=df['crime_type'].value_counts().index)
plt.title("Top 10 Crime Types")
plt.savefig("C:/Users/BHATTI Computers/Desktop/smart_crime_prediction/reports/figures/top_crime_types.png")
plt.show()

# Crime by area
plt.figure(figsize=(8,6))
sns.countplot(data=df, y="area", order=df['area'].value_counts().index)
plt.title("Crimes by Area")
plt.savefig("C:/Users/BHATTI Computers/Desktop/smart_crime_prediction/reports/figures/crimes_by_area.png")
plt.show()

# Arrest distribution
plt.figure(figsize=(6,4))
sns.countplot(x="arrest_made", data=df)
plt.title("Arrest Made Distribution")
plt.savefig("C:/Users/BHATTI Computers/Desktop/smart_crime_prediction/reports/figures/arrest_distribution.png")
plt.show()
```
